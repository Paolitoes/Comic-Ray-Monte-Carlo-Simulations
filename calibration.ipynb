```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#time scale is at 30 seconds, threshold at 108mv, lower PMT
voltages = [1300,1400,1500,1550,1600,1650,1700,1750,1800,1850,1900,1950,2000]
counts = [51,703,1171,1294,1411,1448,1517,1570,1602,1692,1787,2117,2349]
plt.plot(voltages,counts,marker='.',linestyle="")
plt.xlabel('Voltage(V)')
plt.ylabel('Counts')
plt.show()
def gauss(x, A, mu, sigma):
    gauss = A*np.exp( -(x - mu)**2 / (2*sigma**2) )
    return gauss
```
