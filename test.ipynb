```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
u3 = np.random.rand(1000)
```

```python
def bisection(func, a, b, iter):
    m = 0
    for _ in range(iter):
        m = 0.5*(a+b)
        fm = func(m)
        if fm*func(a)>0:
            a=m
        else:
            b=m
    return m

def func(x):
    return x/2 +np.sin(2*x)/2

k = np.zeros_like(u3)

for i,u in enumerate((u3-0.5)*np.pi):
    k[i]=bisection(lambda p: func(p)-u,-2*np.pi,2*np.pi,70)
```

```python
plt.figure()
x4=np.linspace(0,2*np.pi,100)-np.pi
plt.plot(x4,(np.cos(x4)**2),label="cos")
plt.hist(k,bins=30,density=True)
plt.legend()
plt.show()
```
